# File Format Handling

```python
import pandas as pd
from docx import Document
import PyPDF2
```

```python
import win32com.client

resume_path = r"C:\Users\manas\Downloads\P643_DATASET\Resumes"
word = win32com.client.Dispatch("Word.Application")
word.Visible = False
for root, dirs, files in os.walk(resume_path):
    for file in files:
        if file.endswith(".doc") and not file.endswith(".docx"):
            doc_path = os.path.join(root, file)
            docx_path = doc_path + "x"    
            try:
                doc = word.Documents.Open(doc_path)
                doc.SaveAs(docx_path, FileFormat=16) 
                doc.Close()
                print("Converted:", doc_path)
            except Exception as e:
                print("Failed:", doc_path)
                print(e)
word.Quit()
print("Conversion completed.")
```

```python
resume_path = r"C:\Users\manas\Downloads\P643_DATASET\Resumes"

texts = []
categories = []

def extract_docx(file_path):
    try:
        doc = Document(file_path)
        return " ".join([para.text for para in doc.paragraphs])
    except Exception as e:
        print("Error reading DOCX:", file_path)
        print(e)
        return ""

def extract_pdf(file_path):
    text = ""
    try:
        with open(file_path, "rb") as file:
            reader = PyPDF2.PdfReader(file)
            for page in reader.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text
        return text
    except Exception as e:
        print("Error reading PDF:", file_path)
        print(e)
        return ""

for item in os.listdir(resume_path):
    full_path = os.path.join(resume_path, item)
    if os.path.isdir(full_path):
        category = item
        for file in os.listdir(full_path):
            file_path = os.path.join(full_path, file)
            if file.endswith(".docx"):
                text = extract_docx(file_path)
            elif file.endswith(".pdf"):
                text = extract_pdf(file_path)
            else:
                continue
            if text.strip() != "":
                texts.append(text)
                categories.append(category)
    elif os.path.isfile(full_path):
        if item.endswith(".docx"):
            text = extract_docx(full_path)
        elif item.endswith(".pdf"):
            text = extract_pdf(full_path)
        else:
            continue
        if text.strip() != "":
            category = item.split("_")[0]
            texts.append(text)
            categories.append(category)
```

```python
df = pd.DataFrame({
    "Text": texts,
    "Category": categories
})
```

```python
def outlier_capping(df,column):
    Q1= df[column].quantile(0.25)
    Q3= df[column].quantile(0.75)
    IQR= Q3-Q1
    lower_extreme= Q1-1.5*IQR
    upper_extreme= Q3+1.5*IQR
    df[column]= df[column].apply(lambda x: lower_extreme if x<lower_extreme else upper_extreme if x>upper_extreme else x)
    print(f"{column} capped using IQR method.")
```

# Text Preprocessing

```python
# Converting into lower case
df['clean_text'] = df['Text'].str.lower()
```

```python
# Remove Special Characters & Numbers
import re
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
df['clean_text'] = df['clean_text'].apply(clean_text)
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x: " ".join(x.split()))
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words("english"))
def remove_stopwords(text):
    words = text.split()
    return " ".join([word for word in words if word not in stop_words])
df["clean_text"] = df["clean_text"].apply(remove_stopwords)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    words = text.split()
    return " ".join([lemmatizer.lemmatize(wd) for wd in words])
df['clean_text'] = df['clean_text'].apply(lemmatize_text)
```
